```python
from bs4 import BeautifulSoup as bs
import requests 
import time 
```

```python
#this is to search for website pages "for looping from 1 to n "
i = 0
url = "https://wuzzuf.net/search/jobs/?a=hpb&q=&start=" + str(i)
job_results = []
#this to avoid connections errors
#this to request web page and handeling http errors
with requests.Session() as session:
#here you loop for 1000 page in wuzzaf
    while i < 1000:
        i += 1
        url = "https://wuzzuf.net/search/jobs/?a=hpb&q=&start=" + str(i)
        try:
            page = session.get(url, timeout=5)  # Set a timeout value (in seconds) for the request
            page.raise_for_status()  # Raise an exception if the response status is not 200
        except (requests.exceptions.RequestException, requests.exceptions.Timeout) as e:
            print(f"Error occurred while requesting page {url}: {e}")
            continue
#using beautiful soup to parse the html page
#Scraping start from here      
        soup = bs(page.text, 'lxml')
        jobs = soup.find_all('div', 'css-1gatmva e1v1l3u10')
        
        for job in jobs:
            job_title = job.find('h2', class_='css-m604qf').text.strip()
            #this line is to filter jobs title
            if 'Data Analyst' in job_title or 'Data Engineer' in job_title:
                company_name = job.find('div', class_='css-d7j1kk').text.strip().split()
                company_link_element = job.find('div', class_='css-d7j1kk').find('a')
                company_link = company_link_element['href'] if company_link_element and 'href' in company_link_element.attrs else None
                company_loc = job.find('span', class_='css-5wys0k').text.strip().split()
#this to add results in list as dict
                job_info = {
                    'Job title': job_title,
                    'Company name': company_name[0],
                    'Company Link': company_link,
                    'Company Location': f'{company_loc[0]} {company_loc[1]}'
                }
                job_results.append(job_info)

    # Print the job results in batches
    batch_size = 10
    for idx in range(0, len(job_results), batch_size):
        batch = job_results[idx:idx + batch_size]
        print("Batch Results:")
        for job_info in batch:
            print(f'Job title        : {job_info["Job title"]}')
            print(f'Company name     : {job_info["Company name"]}')
            print(f'Company Link     : {job_info["Company Link"]}')
            print(f'Company Location : {job_info["Company Location"]}')
            print('')
        print("--------------------------------------------------")

print("done")
```
